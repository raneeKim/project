```python
merged_df = pd.merge(df1, df2, on='국가명')
merged_df
print(merged_df)
merged_df.to_csv('merged_df.csv',header=True,index=False)
```

```python
df3= pd.read_csv("merged_df.csv")
print(df3)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import QuantileTransformer

# 데이터 준비
df_f = df3[['수출건수', '수출금액', '수입건수', '수입금액', '무역수지', '국내총생산', '1인당 총생산']]

# 정규화
qt = QuantileTransformer()
cc_scaled = qt.fit_transform(df_f)

# 클러스터링
kmeans = KMeans(n_clusters=3, random_state=10)
clusters = kmeans.fit(cc_scaled)

# 군집 결과를 원본 데이터에 추가
cc['cluster'] = clusters.labels_

# 군집별 데이터 수 확인
print(cc.groupby('cluster').count())

# 산점도 시각화
colors = ['red', 'green', 'blue']
plt.scatter(x=cc['cluster'], y=cc['무역수지'], c=cc['cluster'], cmap='viridis')
plt.xlabel('Cluster')
plt.ylabel('무역수지')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import QuantileTransformer

# 데이터 준비 (PCA)?
df_f = df3[['수출건수', '수출금액', '수입건수', '수입금액', '무역수지', '국내총생산', '1인당 총생산']]

# 정규화
qt = QuantileTransformer()
cc_scaled = qt.fit_transform(df_f)

# 클러스터링
kmeans = KMeans(n_clusters=3, random_state=10)
clusters = kmeans.fit(cc_scaled)

# 군집 결과를 원본 데이터에 추가
cc['cluster'] = clusters.labels_

# 군집별 데이터 수 확인
print(cc.groupby('cluster').count())

# 군집별 국가명 출력
for cluster_id in range(3):
    cluster_countries = df3.loc[cc['cluster'] == cluster_id, '국가명']
    print(f'Cluster {cluster_id}:')
    print(cluster_countries)

# 산점도 시각화
plt.scatter(x=cc['cluster'], y=cc['무역수지'], c=cc['cluster'], cmap='viridis')
plt.xlabel('Cluster')
plt.ylabel('무역수지')
plt.show()
```
