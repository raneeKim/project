```python
df_member['연령'].describe
```

```python
def age_age(row) :
    if row >= 44 :
        return '40대중반'
    elif row < 44 and row >= 40 :
        return '40대초반' 
    elif row < 40 and row >= 37 :
        return '30대후반'
    elif row < 37 and row >= 34 :
        return '30대중반'
    elif row < 34 and row >= 30 :
        return '30대초반'
    elif row < 30 and row >= 27 :
        return '20대후반'
    elif row < 27 and row >= 24 :
        return '20대중반'
    else : return '20대초반'

df_member['연령대'] = df_member['연령'].apply(age_age)
df_member['연령대'].value_counts()
```

```python
ct5=  pd.crosstab(df_member['연령대'], df_member['더블구매유무'])
print(ct5)
stats.chi2_contingency(ct5)
```
